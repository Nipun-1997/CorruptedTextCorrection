<h3> Preparation of Data </h3>

Since our data is in txt format which contains SMS text in one line, Standard English in second line and Chinese Translation of standard English in 3rd line. We would be using only SMS_Text and Standard English for our problem Statement.
dataset details.PNG

```
"""In this function we would be sending our file From the above image, sms text is the first line along
 with its english text followed by its chinese translation. Here with the help of index text has been 
 seperated.
 Logic: if index is divisible by 3 then that line is smstext and append into smstext list
        if remainder of index is 1 then that line is english text and append into normal_list
        if remainder of index is 2 then that line is chinese text
 return sms_text and normal_text list"""
def preparing_dataset(file_name):
    files=open(file_name,'r')
    lines=files.readlines()
    files.close()
    sms_text=[]
    normal_text=[]

    for index,line in enumerate(lines):
        if index%3==0:
           sms_text.append(line.strip())
        elif index%3==1:
            normal_text.append(line.strip())
        else:
            pass
    return sms_text,normal_text             

```

```
document_name='/content/en2cn-2k.en2nen2cn'
sms_text,normal_text=preparing_dataset(document_name)
print(len(sms_text))
print(len(normal_text))
print("SMS Text:",sms_text[1011])
print("Normal Text:",normal_text[1011])
```

```
"""Creating dataframe from the sms_text and english_text"""
data_preparation={'SMS_TEXT':sms_text,
                  'ENGLISH_TEXT': normal_text}
```

```
data=pd.DataFrame(data_preparation)
print(data.head(10))
print(data.tail(10))
```
